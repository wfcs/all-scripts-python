```python
import requests
import pandas as pd
```

```python
# Função para realizar requisições paginadas e incrementais
def fetch_data():
    url = "https://api-analytics.checklistfacil.com.br/"
    endpoint = "v1/evaluations/results?"
    
    # Parâmetros iniciais
    query_params = {
        "checklistId": 592565,  # ID do checklist que você está buscando
    }

    headers = {
        "Authorization": "Bearer QxMtxjfyjlOdQZ2Ca2QeEK9ftEsptv21RdMGYpu40R2s8FMBCI1VtdfCymwb5U9AKu5QEXoo3TTssIe8PFcLCS0dO1ekHT9GQ1t3dmi3zvEJiFbTgKRuhFqYQ2Ez6luv", 
        "Content-Type": "application/json"
    }

    all_data = []  # Lista para armazenar todos os dados

    page = 1  # Começa da primeira página
    # pivot = None  # Para armazenar o valor de pivot para paginação incremental

    while True:
        # Atualiza os parâmetros com a página e o pivot (se disponível)
        query_params["page"] = page

        # Fazendo a requisição
        response = requests.get(url + endpoint, params=query_params, headers=headers)
        
        # Verificando se a requisição foi bem-sucedida
        if response.status_code == 200:
            data = response.json().get('data', [])
            
            # Se não houver dados, significa que chegamos ao fim
            if not data:
                break

            # Adiciona os dados obtidos à lista
            all_data.extend(data)

            # Atualiza o valor de pivot para a próxima requisição (usando o último item)
            pivot = data[-1].get('evaluationId')  # Considera o ID da última avaliação como pivot para a próxima página

            # Incrementa a página
            page += 1

        else:
            print(f"Erro na requisição: {response.status_code}")
            break
    
    # Retornando todos os dados
    return pd.DataFrame(all_data)
```

```python
# Coletando os dados
df = fetch_data()
```

```python
# Verificando se os dados foram coletados
if not df.empty:
    # Tratando os dados
    df = df.drop_duplicates(subset=['evaluationId'])

    # Convertendo os tipos das colunas
    df['evaluationId'] = df['evaluationId'].astype('int64')
    df['status'] = df['status'].astype('int64')
    df['score'] = pd.to_numeric(df['score'], errors='coerce')
    df['unitId'] = df['unitId'].astype('int64')
    df['userId'] = df['userId'].astype('int64')
    df['startedAt'] = pd.to_datetime(df['startedAt'], errors='coerce')
    df['concludedAt'] = pd.to_datetime(df['concludedAt'], errors='coerce')
    df['approvedAt'] = pd.to_datetime(df['approvedAt'], errors='coerce')
    df['itemId'] = df['itemId'].astype('int64')
    df['categoryId'] = df['categoryId'].astype('int64')
    df['resultId'] = df['resultId'].astype('int64')
    df['scaleId'] = df['scaleId'].astype('int64')
    df['answeredAt'] = pd.to_datetime(df['answeredAt'], errors='coerce')
    df['updatedAt'] = pd.to_datetime(df['updatedAt'], errors='coerce')
    df['createdAt'] = pd.to_datetime(df['createdAt'], errors='coerce')
    df['initialLatitude'] = pd.to_numeric(df['initialLatitude'], errors='coerce')
    df['finalLongitude'] = pd.to_numeric(df['finalLongitude'], errors='coerce')

    # Exibindo os dados
    print(df)
else:
    print("Nenhum dado foi retornado.")
```
